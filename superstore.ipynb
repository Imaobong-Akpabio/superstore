##5. Are the division of "36 and 6" the same as their floor division? If no, state your reason.

The asnswer is No, because when using the division operator the answer will always return the quotient as a float whereas when using the floor division,the answer will always return the value of the floor of the quotient, without returning a float value.

######Please refer to the example below:

```python
#Using the division operator
36/6
# The division operator returns the quotient as a float.
#The result include the decimal
```

```python
#Using the floor division
36//6
#The floor division return the quotient without returning the float value
# Any decimal place is discarded
```

##8. Can the subtraction operator act on a string value?

No, the subtraction operator (-) cannot directly act on string values in Python. Attempting to subtract one string from another will result in a TypeError.
######Please refer to the example below

```python
x= "hello"
y= "world"
result = x- y
print(result)
# This will raise a TypeError
```

```python
x= "hello"
y= "world"
result = x+" "+ y
print(result)
```
